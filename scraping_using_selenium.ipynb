```python
import pandas as pd
```

```python
df_historical_data = pd.read_csv('FIFA_World_Cup_Historical_Data.csv')
df_fixture = pd.read_csv('FIFA_World_Cup_Fixtures_Data.csv')
```

```python
df_fixture['HomeTeam'] = df_fixture['HomeTeam'].str.strip()
df_fixture['AwayTeam'] = df_fixture['AwayTeam'].str.strip()
```

```python
# Deleting matches with walkover

delete_index = df_historical_data[df_historical_data['HomeTeam'].str.contains('Sweden') & df_historical_data['AwayTeam'].str.contains('Austria')].index
df_historical_data.drop(index = delete_index, inplace = True)
```

```python
df_historical_data['Score'] = df_historical_data['Score'].str.replace('[(a.e.t.)]', '', regex = True)
```

```python
#Cleaning blank spaces from HomeTeam, Score and AwayTeam columns

df_historical_data['HomeTeam'] = df_historical_data['HomeTeam'].str.strip()
df_historical_data['AwayTeam'] = df_historical_data['AwayTeam'].str.strip()
```

```python
# Splitting score columns into Home and Away goals and dropping the score column

df_historical_data[['HomeGoals', 'AwayGoals']] = df_historical_data['Score'].str.split('–', expand = True)
df_historical_data.drop('Score', axis = 1, inplace = True)
```

```python
df_historical_data
```

```python
# Changing Datatypes

df_historical_data = df_historical_data.astype({'HomeGoals': int, 'AwayGoals': int, 'Year': int})
```

```python
# Creating new column "TotalGoals"

df_historical_data['TotalGoals'] = df_historical_data['HomeGoals'] + df_historical_data['AwayGoals']
```

```python
df_historical_data
```

```python
df_historical_data.to_csv('Clean_FIFA_World_Cup_Historical_Data.csv', index = False)
df_fixture.to_csv('Clean_FIFA_World_Cup_Fixtures_Data.csv', index = False)
```
